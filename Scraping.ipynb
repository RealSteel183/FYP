```python
from bs4 import BeautifulSoup
import os
```

```python
# Path to the directory containing your files
directory_path = 'data/aspxAl/'

# Pattern to look for in links
link_pattern = "/search/DataSheet.aspx?MatGUID"

# Words to exclude in link text
exclude_words = ["Overview of materials", "Composition Spec","discontinued" ,"Thermal Transfer Composites", "Specification", "Grade", ""]

# Initialize an empty list to store all matching links
all_matching_links = []

# Filenames of the ASPX files to process
file_names = [
    "1000Al.aspx", "2000Al.aspx", "3000Al.aspx", "4000Al.aspx",
    "5000Al1.aspx", "5000Al2.aspx", "6000Al.aspx", "7000Al.aspx", "CastAlloy1.aspx", "CastAlloy2.aspx", "CastAlloy3.aspx"
]

# Iterate over each file, parse it, and extract the required links
for file_name in file_names:
    file_path = os.path.join(directory_path, file_name)
    
    with open(file_path, 'r') as file:
        html_content = file.read()
        soup = BeautifulSoup(html_content, 'html.parser')
        links = soup.find_all('a', href=True)
        
        for link in links:
            # Check if link contains the pattern and does not contain any of the exclude words
            if link_pattern in link['href'] and not any(exclude_word in link.text for exclude_word in exclude_words):
                all_matching_links.append(link['href'])

# Optionally, print the links or save them to a file
print("Found links:")
for link in all_matching_links:
    print(link)

# Saving the links to a text file
with open('data/extractedLinksAl.txt', 'w') as f:
    for link in all_matching_links:
        f.write("%s\n" % link)

print("Links have been saved to extractedLinksAl.txt")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Base URL to prepend to each link
base_url = "https://www.matweb.com"

# Path to the file containing the links
links_file_path = 'data/extractedLinks.txt'

# Read the links from the file
with open(links_file_path, 'r') as file:
    links = [line.strip() for line in file if line.strip()]

# Initialize the WebDriver with ChromeDriverManager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Base directory where the HTML files will be saved
base_dir = 'data/pagesAl/'

for index, link in enumerate(links[579:], start=614):  # Start counting from 1
    # Prepend the base URL to the link
    full_url = base_url + link
    print(full_url)
    # Navigate to the URL
    driver.get(full_url)
    
    time.sleep(5)
    
    # Get the page source
    html_content = driver.page_source
    
    # Generate a unique filename for each page
    file_path = f'{base_dir}page{index}.html'
    
    # Write the HTML content to a file
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(html_content)
    
    print(f'Page HTML has been saved to {file_path}')

# Close the browser
driver.quit()
```
